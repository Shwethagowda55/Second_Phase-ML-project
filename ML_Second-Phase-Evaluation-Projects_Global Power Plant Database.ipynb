```python
df = df.drop(['country','wepp_id','url','country_long','geolocation_source','name','gppd_idnr'],axis = 1)
```

```python
df['total_generation'] = df['generation_gwh_2013'] + df['generation_gwh_2014'] + df['generation_gwh_2015'] + df['generation_gwh_2016'] + df['generation_gwh_2017']+df['generation_gwh_2018']+df['generation_gwh_2019'] 
```

```python
df = df.drop(['generation_gwh_2013','generation_gwh_2014','generation_gwh_2015','generation_gwh_2016','generation_gwh_2017','generation_gwh_2018','generation_gwh_2019'],axis = 1)
```

```python
df.head()
```

```python
df.info()
```

```python
# missing values
df.isnull().sum()
```

```python
df['latitude'].fillna(df['latitude'].mode()[0], inplace= True)
df['longitude'].fillna(df['longitude'].mode()[0], inplace= True)
df['other_fuel1'].fillna(df['other_fuel1'].mode()[0], inplace= True)
df['commissioning_year'].fillna(df['commissioning_year'].mode()[0], inplace= True)
df['owner'].fillna(df['owner'].mode()[0], inplace= True)
```

```python
df['year_of_capacity_data'].fillna(df['year_of_capacity_data'].mode()[0], inplace= True)
df['generation_data_source'].fillna(df['generation_data_source'].mode()[0], inplace= True)
```

```python
df['other_fuel2'].fillna(df['other_fuel2'].mode()[0], inplace= True)
```
